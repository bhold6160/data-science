```python
import pandas as pd
import numpy as np
from statistics import median
```

```python
df = pd.read_csv('./vgsales.csv')
df.head()

#What are the top 20 highest grossing
```

```python
#Which company is the most common publisher
pd.DataFrame(df['Publisher']).groupby(df.Publisher).count().sort_values('Publisher')
```

```python
#What's the most common platform
pd.DataFrame(df['Platform']).groupby(df.Platform).count().sort_values('Platform')
```

```python
#What about the most common genre?
pd.DataFrame(df['Genre']).groupby(df.Genre).count().sort_values('Genre')
```

```python
#What are the top 20 highest grossing games?
pd.DataFrame(df[['Name', 'Global_Sales']]).head(21)
```

```python
#For North American video game sales, what’s the median?
df[(df['NA_Sales']< 0.085) & (df['NA_Sales'] > 0.075)].head(11)
```

```python
#For the top-selling game of all time, how many standard deviations above/below the mean are its sales for North America?
top_game = df[['Name', 'Global_Sales', 'NA_Sales']].sort_values('Global_Sales', ascending = False)[:1]
american_sales_std = df['NA_Sales'].std()
american_sales_mean = df['NA_Sales'].mean()
(top_game['NA_Sales'] - american_sales_mean) / american_sales_std
```

```python
#The Nintendo Wii seems to have outdone itself with games. How does its average number of sales compare with all of the other platforms?
pd.DataFrame(df[['Platform', 'Global_Sales']]).head(1)

pd.DataFrame(df[['Platform', 'Global_Sales']]).groupby(df.Platform).count().sort_values('Platform', ascending=False)
```
